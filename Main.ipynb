```python
import random

maleFirstNames = []
femaleFirstName = []
lastName = []

malename = open("male-first-names.txt", "r")
maleFirstNames = malename.read().splitlines()
ranMale = random.choice(maleFirstNames)

female = open("female-first-names-1.txt", "r")
femaleFirstName = female.read().splitlines()
ranFemale = random.choice(femaleFirstName)

last = open("last-names.txt", "r")
lastName = last.read().splitlines()
ranLast = random.choice(lastName)

gender = random.choice(["male", "female"])

if gender == "male":
    firstName = random.choice(maleFirstNames)
else:
    firstName = random.choice(femaleFirstName)

ranLast = random.choice(lastName)

fullName = firstName + " " + ranLast

startingYear = 2019
endingYear = 2029
Year = random.randint(startingYear, endingYear)

courses = [
    "MAT112", "CPE123", "GST104", "MAT221", "CPI109", "CPT342", "EET124", "AET214"
]

Grades = {"A": 5, "B": 4, "C": 3, "D": 2, "E": 1, "F": 0}

# Initialize dictionaries to store course grades and credit units
course_grades = {}
course_units = {}

# Randomly assign grades to each course
for course in courses:
    course_grades[course] = random.choice(list(Grades.keys()))
    course_units[course] = random.randint(1, 3)  # Assign a random credit unit between 1 and 3

# Calculate total credit load and total credit units
total_credit_load = sum(Grades[course_grades[course]] * course_units[course] for course in courses)
total_credit_units = sum(course_units[course] for course in courses)

# Calculate GPA based on the provided formula
gpa = total_credit_load / total_credit_units if total_credit_units > 0 else 0  # Avoid division by zero


print("Name: " + fullName)
print("\n")
print("Gender: " + gender.capitalize())
print("\n")
print("Year:", Year)
print("\n")
print("========  =====  ============\nCOURSE    GRADE   CREDIT UNIT\n========  =====  ============")
for course in courses:
    print(f"{course}\t  {course_grades[course]}\t            {course_units[course]}")
print("========  =======  ===========")
print("\n")
print(f"TOTAL CREDIT LOAD: {total_credit_units}")  # Fixed total credit load
print("\n")
print(f"GPA: {gpa:.2f}")  # Display GPA with 2 decimal places
print("\n")
if gpa < 2.49:
    remark = "Poor"
elif gpa >2.50 <3.49 :
    remark = "Good"
elif gpa >3.50 < 4.49:
    remark = "Very Good"
else:
    remark = "Excellent"
print("REMARK: ", remark)
```
